добавление полей для aif

```python
import json
import os

# Пути к файлам
input_file = '/content/aif_thesis_master.json'
output_file = '/content/aif_thesis_master_updated.json'

def transform_aif_data(input_path, output_path):
    if not os.path.exists(input_path):
        print(f"Файл {input_path} не найден.")
        return

    with open(input_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    updated_data = []

    for index, item in enumerate(data):
        # Создаем новый объект с нужным порядком полей
        transformed_item = {
            "url": item.get("url", ""),
            "source": "aif.ru",                 # Указываем источник для этого файла
            "language": "ru",                   # Указываем язык
            "category": item.get("category", ""),
            "headline": item.get("headline", ""),
            "text": item.get("text", ""),
            "date": item.get("date", "")        # Поле date уже есть в исходном файле
        }
        updated_data.append(transformed_item)

    # Сохраняем обновленный файл
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(updated_data, f, ensure_ascii=False, indent=2)

    print(f"Обновлено записей: {len(updated_data)}")
    print(f"Файл сохранен по пути: {output_path}")

# Запуск трансформации
transform_aif_data(input_file, output_file)
```

очистка от \n

```python
import json
import os

input_path = '/content/eg_thesis_master_updated.json'
output_path = '/content/eg_thesis_master_updated_cleaned.json'

def clean_eg_data(input_file, output_file):
    if not os.path.exists(input_file):
        print(f"Файл {input_file} не найден.")
        return

    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    for item in data:
        if "headline" in item and item["headline"]:
            # 1. strip() убирает \n и обычные пробелы по краям
            # 2. replace('\xa0', ' ') заменяет неразрывные пробелы на обычные
            # 3. Снова strip() для финальной очистки
            cleaned_headline = item["headline"].replace('\xa0', ' ').strip()
            item["headline"] = cleaned_headline

    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

    print(f"Очистка завершена. Файл сохранен как: {output_file}")

# Запуск
clean_eg_data(input_path, output_path)
```

обьединение данных.

```python
import json
import os

# Список путей к вашим обновленным файлам
files_to_merge = [
    '/content/aif_thesis_master_updated.json',
    '/content/eg_thesis_master_updated_cleaned.json',
    '/content/lenta_thesis_master_updated.json',
    '/content/mail_thesis_master_updated.json'
]

combined_data = []

def merge_datasets(file_list, output_path):
    for file_path in file_list:
        if os.path.exists(file_path):
            with open(file_path, 'r', encoding='utf-8') as f:
                try:
                    data = json.load(f)
                    # Если данные в файле — это список, добавляем его элементы в общий список
                    if isinstance(data, list):
                        combined_data.extend(data)
                    else:
                        combined_data.append(data)
                    print(f"Файл {file_path} успешно добавлен. Записей: {len(data)}")
                except Exception as e:
                    print(f"Ошибка при чтении файла {file_path}: {e}")
        else:
            print(f"Предупреждение: Файл {file_path} не найден.")

    # Сохраняем итоговый результат
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(combined_data, f, ensure_ascii=False, indent=2)

    print("-" * 30)
    print(f"Объединение завершено. Итого задействовано файлов: {len(file_list)}")
    print(f"Общее количество новостей в итоговом файле: {len(combined_data)}")
    print(f"Путь к файлу: {output_path}")

# Запуск
merge_datasets(files_to_merge, '/content/dataset_2424.json')
```

проверка категории и источника

```python
import pandas as pd
import json

# Загружаем данные
file_path = '/content/dataset_2427.json'

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Создаем DataFrame
df = pd.DataFrame(data)

# Группируем по источнику (source) и категории (category)
# .size() считает количество, .unstack() делает таблицу наглядной
pivot_table = df.groupby(['source', 'category']).size().unstack(fill_value=0)

print("### Сводная таблица: Источник vs Категория")
print(pivot_table)

print("\n" + "="*50 + "\n")

# Если нужно вывести просто списком (топ категорий по каждому источнику)
print("### Детальный список:")
counts = df.groupby(['source', 'category']).size()
print(counts)

# Общая статистика по категориям (без привязки к источнику)
print("\n" + "="*50 + "\n")
print("### Общий подсчет по всем категориям:")
print(df['category'].value_counts())
```

обьединим категории

```python
import pandas as pd
import json

# 1. Загрузка
file_path = '/content/dataset_2427.json'
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)
df = pd.DataFrame(data)

# 2. Словарь переименования (ключ - что было, значение - что станет)
unification_map = {
    # Группа "Общество"
    'Общество': 'Общество',
    'Россия': 'Общество',
    'Новости': 'Общество',
    'Главное': 'Общество',

    # Группа "Политика и Безопасность"
    'Политика': 'Политика и Безопасность',
    'СВО': 'Политика и Безопасность',
    'Силовые структуры': 'Политика и Безопасность',
    'Мир': 'Политика и Безопасность',

    # Группа "Технологии"
    'Наука и техника': 'Технологии и СМИ',
    'Интернет и СМИ': 'Технологии и СМИ',

    # Остальные оставляем или переименовываем для красоты
    'Экономика': 'Экономика',
    'Происшествия': 'Происшествия',
    'Шоу-бизнес': 'Шоу-бизнес'
}

# 3. Применяем изменения
df['category'] = df['category'].map(unification_map)

# 4. Проверяем результат
print("Новое распределение категорий:")
print(df['category'].value_counts())

# 5. Сохраняем в новый JSON
final_json = df.to_dict(orient='records')
with open('/content/dataset_final_cleaned.json', 'w', encoding='utf-8') as f:
    json.dump(final_json, f, ensure_ascii=False, indent=2)

print("\nФайл 'dataset_final_cleaned.json' готов!")
```
